# Easy problems

```python
# Check age and status
# Statement: Return True if age is between 18 and 65 (inclusive) AND has_license is True.
# Input: Integer age, boolean has_license
# Output: Boolean
# Example:
# Input: 25, True → Output: True
# Input: 17, True → Output: False
# Input: 25, False → Output: False

def age_status(age, has_license =True):
    if (age >=18 and age <= 65) and has_license:
        return True
    return False

age_status(25, has_license =False)
```

```python
# Valid password check
# Statement: Check if password length is at least 8 AND contains at least one digit.
# Input: String password
# Output: Boolean
# Example:
# Input: "Password123" → Output: True
# Input: "short" → Output: False
# Input: "nopassword" → Output: False

def pass_check(password):
    if len(password) >= 8 and (any(char.isdigit() for char in password)):
        return True
    return False

pass_check("Password0")
```

```python
# Weekend or holiday
# Statement: Return True if it's a weekend (Saturday=6, Sunday=7) OR is_holiday is True.
# Input: Integer day (1-7), boolean is_holiday
# Output: Boolean
# Example:
# Input: 6, False → Output: True
# Input: 3, True → Output: True
# Input: 3, False → Output: False

def weekend_holiday(day, is_holiday= False):
    if day == 6 or day==7 or is_holiday is True:
        return True
    return False

weekend_holiday(1, is_holiday= True)
```

```python
# Not empty and positive
# Statement: Return True if a number is positive AND a list is not empty.
# Input: Integer num, list arr
# Output: Boolean
# Example:
# Input: 5, [1, 2, 3] → Output: True
# Input: -5, [1, 2] → Output: False
# Input: 5, [] → Output: False

def not_empty_and_positive(num, lst):
    if num > 0 and len(lst) > 0:
        return True
    return False

not_empty_and_positive(5, [1])
```

# Medium problems 

```python
# Complex eligibility check
# Statement: Person is eligible if (age >= 18 AND has_id) OR (age >= 16 AND has_parent_consent).
# Input: Integer age, boolean has_id, boolean has_parent_consent
# Output: Boolean
# Example:
# Input: 20, True, False → Output: True
# Input: 16, False, True → Output: True
# Input: 15, True, True → Output: False

def eligibility(age: int, has_id: bool, has_parent_consent: bool) -> bool:
    if (age >= 18 and has_id) or (age >= 16 and has_parent_consent):
        return True
    return False

eligibility(15, True, True)
```

```python
# Safe division with checks
# Statement: Return True if division is safe (divisor != 0 AND dividend is not None).
# Input: Number dividend, number divisor
# Output: Boolean
# Example:
# Input: 10, 2 → Output: True
# Input: 10, 0 → Output: False
# Input: None, 5 → Output: False

def division_check(dividend: int | None, divisor:int) -> bool:
    if (dividend is not None and divisor != 0):
        return True
    return False

division_check(None, 5)
```
