# Easy problems

```python
# Grade calculator
# Statement: Given a score (0–100), assign a letter grade: A (90–100), B (80–89), C (70–79), D (60–69), F (<60).
# Input: Single integer score
# Output: Single character (A, B, C, D, or F)
# Example:
# Input: 85 → Output: B
# Input: 45 → Output: F

def grade(score):
    if score >=90 and score <= 100:
        return "A"
    elif score >=80 and score <= 89:
        return "B"
    elif score >=70 and score <= 79:
        return "C"
    elif score >=60 and score <= 69:
        return "D"
    else:
        return "F"

grade(score = 99)  
```

```python
# Positive, negative, or zero
# Statement: Determine if a number is positive, negative, or zero.
# Input: Single integer n
# Output: String: "Positive", "Negative", or "Zero"
# Example:
# Input: -5 → Output: Negative
# Input: 0 → Output: Zero

def pos_neg_zero(num):
    if num == 0:
        return "Zero"
    if num < 0:
        return "Negative"
    if num > 0:
        return "Positive"

pos_neg_zero(num = 5)
```

```python
# Leap year checker
# Statement: Determine if a year is a leap year. 
# Rules: divisible by 4, but not by 100 unless also divisible by 400.
# Input: Single integer year
# Output: Boolean True or False
# Example:
# Input: 2024 → Output: True
# Input: 1900 → Output: False
# Input: 2000 → Output: True

def leap_year(year):
    if year % 4 == 0 and (year % 100 != 0):
        return True
    else:
        return False

leap_year(year = 1900)
```

```python
# Triangle type
# Statement: Given three side lengths, 
# determine if they form a valid triangle and, if so, classify as "Equilateral", "Isosceles", or "Scalene". 
# If invalid, return "Invalid".
# Input: Three integers a, b, c
# Output: String
# Example:
# Input: 3, 3, 3 → Output: Equilateral
# Input: 3, 4, 5 → Output: Scalene
# Input: 1, 2, 5 → Output: Invalid

def triangle(a, b, c):
    if a == b == c:
        return "Equilateral"
    elif (a != b != c) and (a + b) > c:
        return "Scalene"
    elif (a == b != c) and (a + b) > c:
        return "Isosceles"
    else:
        return "Invalid"

triangle(3, 3, 5)
    
```

```python
# Day of week
# Statement: Given a number (1–7), return the day name: 1=Monday, 2=Tuesday, ..., 7=Sunday.
# Input: Single integer day (1–7)
# Output: String day name
# Example:
# Input: 3 → Output: Wednesday
# Input: 7 → Output: Sunday

def dow(num):
    if num == 1:
        return "Monday"
    elif num == 2:
        return "Tuesday"
    elif num ==3:
        return "Wednesday"
    elif num ==4:
        return "Thursday"
    elif num ==5:
        return "Friday"
    elif num ==6:
        return "Saturday"
    elif num ==7:
        return "Sunday"
    else:
        return "Not a day"

dow(10)
```

```python
# Quadratic equation solver
# Statement: Given coefficients a, b, c of ax² + bx + c = 0, 
# determine the number of real roots (0, 1, or 2). If a == 0, return "Not a quadratic".
# Input: Three integers a, b, c
# Output: Integer (0, 1, or 2) or string "Not a quadratic"
# Example:
# Input: 1, -5, 6 → Output: 2 (roots: 2 and 3)
# Input: 1, -4, 4 → Output: 1 (double root: 2)
# Input: 1, 1, 1 → Output: 0 (no real roots)

# x = -b +- sqrt(b^2 - 4*a*c) / 2a

import math

def quadratic(a, b, c):
    if a==0:
        return "Not a quadratic"
    else:
        d = (b**2) - (4*a*c)
        if (a == b == c) or d == 0:
            return "Output: 0 (no real roots)"
        x1 = (-b + math.sqrt(d)) / 2*a
        x2 = (-b - math.sqrt(d)) / 2*a
        if x1 and x2:
            return f"Output: 2 (roots: {x1} and {x2})"
        elif x1 == x2:
            return f"Output: 1 (roots: {x1})"
        else:
            return "Output: 0 (no real roots)"


quadratic(1, -4, 4 )
```

```python
# Age group classifier
# Statement: Classify a person's age: "Child" (<13), "Teen" (13–19), "Adult" (20–64), "Senior" (≥65).
# Input: Single integer age
# Output: String category
# Example:
# Input: 15 → Output: Teen
# Input: 70 → Output: Senior

def age_group(age):
    if age < 13:
        return "Child"
    elif age >= 13 and age <= 19:
        return "Teen"
    elif age >= 20 and age <= 64:
        return "Teen"    
    elif age >= 65:
        return "Senior"
    else:
        return "Invalid age"
```

```python
# Password strength checker
# Statement: Check password strength: "Weak" (<6 chars), "Medium" (6–10 chars, letters only), "Strong" (>10 chars or contains digits/special chars).
# Input: String password
# Output: String ("Weak", "Medium", or "Strong")
# Example:
# Input: "abc" → Output: Weak
# Input: "password" → Output: Medium
# Input: "Pass123!" → Output: Strong
import re

def password_strength(password):
    if len(password) < 6:
        return "Weak"
    elif len(password) >= 6:
        pattern = r'[^\w\s]|\d'
        if re.search(pattern, password):
            return "Strong"
        else:
            return "Medium"
    else:
        return None

password_strength('Pass123!')
```

```python
# BMI category
# Statement: Calculate BMI = weight / (height²) and classify: "Underweight" (<18.5), "Normal" (18.5–24.9), "Overweight" (25–29.9), "Obese" (≥30).
# Input: Two floats weight (kg), height (meters)
# Output: String category
# Example:
# Input: 70, 1.75 → Output: Normal
# Input: 90, 1.70 → Output: Overweight

def bmi(weight, height):
    bmi = weight / height**2
    if bmi < 18.5:
        return "Underweight"
    elif bmi >= 18.5 and bmi <=24.9:
        return "Normal"
    elif bmi >=25 and bmi <=29.9:
        return "Overweight"
    else:
        return "Obese"

```
