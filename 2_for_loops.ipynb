# Easy problems

```python
# Sum of numbers
# Statement: Calculate the sum of all integers from 1 to n (inclusive).
# Input: Single integer n
# Output: Integer sum
# Example:
# Input: 5 → Output: 15 (1+2+3+4+5)
# Input: 10 → Output: 55

def sum_of_int(num):
    sum = 0
    for i in range(1, num + 1):
        sum += i
    return sum

sum_of_int(5)
```

```python
# Count even numbers
# Statement: Count how many even numbers are in a list.
# Input: List of integers numbers
# Output: Integer count
# Example:
# Input: [1, 2, 3, 4, 5, 6] → Output: 3
# Input: [1, 3, 5] → Output: 0

def cnt_even(nums):
    cnt = 0
    for i in nums:
        if i % 2 == 0:
            cnt += 1
    return cnt

cnt_even([1, 3, 5])
```

```python
# Print multiplication table
# Statement: Print the multiplication table for a number up to 10.
# Input: Single integer n
# Output: Print lines like "n x 1 = result", "n x 2 = result", etc.
# Example:
# Input: 5 → Output: 5 x 1 = 5, 5 x 2 = 10, ..., 5 x 10 = 50

def mul_tab(num):
    for i in range(1,11):
        print(f"{num}X{i} = {num*i}")

mul_tab(5)
```

```python
# Find index of element
# Statement: Find the first index of a target value in a list. 
# Return -1 if not found.
# Input: List arr, integer target
# Output: Integer index or -1
# Example:
# Input: [10, 20, 30, 20], 20 → Output: 1
# Input: [10, 20, 30], 40 → Output: -1

def find_idx(nums, find_num):
    for i, x in enumerate(nums):
        if x == find_num:
            return i
    return -1

find_idx([10, 20, 30, 20], 20)
```

# Medium problems

```python
# Reverse a list
# Statement: Create a new list with elements in reverse order 
# (don't use .reverse() or [::-1]).
# Input: List arr
# Output: New reversed list
# Example:
# Input: [1, 2, 3, 4] → Output: [4, 3, 2, 1]
# Input: ['a', 'b', 'c'] → Output: ['c', 'b', 'a']

def rev_list(lst):
    rev_lst = []
    for i in range(len(lst), 0, -1):
        rev_lst.append(i)
    return rev_lst

rev_list([1, 2, 3, 4])
```

```python
# Pair elements
# Statement: Given two lists of equal length, 
# create a list of tuples pairing corresponding elements.
# Input: Two lists list1, list2
# Output: List of tuples
# Example:
# Input: [1, 2, 3], ['a', 'b', 'c'] → Output: [(1, 'a'), (2, 'b'), (3, 'c')]

def pairs(lst1, lst2):
    new_lst = []
    for x, y in zip(lst1, lst2):
        new_lst.append((x, y))
    return new_lst

pairs([1, 2, 3], ['a', 'b', 'c'])
```

```python
# find all indices
# Statement: Find all indices where a target value appears in a list.
# Input: List arr, integer target
# Output: List of indices
# Example:
# Input: [10, 20, 30, 20, 40, 20], 20 → Output: [1, 3, 5]
# Input: [1, 2, 3], 4 → Output: []

def find_idx(lst, find):
    new_lst = []
    for i, x in enumerate(lst):
        if x == find:
            new_lst.append(i)
    return new_lst

find_idx([10, 20, 30, 20, 40, 20], 20)

    
```

# Hard problems

```python
# Sum of digits
# Statement: Calculate the sum of all digits in a number.
# Input: Single integer n
# Output: Integer sum of digits
# Example:
# Input: 1234 → Output: 10 (1+2+3+4)
# Input: 999 → Output: 27

def sum_digits(num):
    sum = 0
    for i in str(num):
        sum += int(i)
    return sum

sum_digits(1234)
```

```python
# Count vowels in string
# Statement: Count the number of vowels (a, e, i, o, u, case-insensitive) in a string.
# Input: String text
# Output: Integer count
# Example:
# Input: "Hello World" → Output: 3
# Input: "Python" → Output: 1

def cnt_vowels(word):
    cnt = 0
    for i in word:
        if i in "aeiou" or i in "AEIOU":
            cnt += 1
    return cnt

cnt_vowels("Hello World")
```

```python
# Find maximum and its index
# Statement: Find the maximum value in a list and return both the value and its index. 
# If multiple max values, return the first index.
# Input: List arr
# Output: Tuple (max_value, index)
# Example:
# Input: [3, 7, 2, 7, 1] → Output: (7, 1)
# Input: [10] → Output: (10, 0)

def max_val(lst):
    max_ = float('-inf')
    max_idx = []
    lst_ = []
    for i, x in enumerate(lst):
        if max_ <= x:
            max_ = x
            max_idx = i
    lst_.append((max_, max_idx))

    return lst_[0]

max_val([3, 7, 2, 7, 1])
```

```python
# Alternating sum
# Statement: Calculate alternating sum: first element - second + third - fourth + ...
# Input: List of numbers arr
# Output: Integer/float result
# Example:
# Input: [1, 2, 3, 4, 5] → Output: 3 (1-2+3-4+5)
# Input: [10, 5] → Output: 5

def alternate_sum(lst):
    total_sum = 0
    total_sub = 0
    for i, x in enumerate(lst):
        if i % 2 == 0:
            total_sum += x
        else:
            total_sub += x
    return total_sum - total_sub

alternate_sum([10, 5])
```

```python
# Compare lists element-wise
# Statement: Compare two lists element-wise. 
# Return a list of "Equal" or "Different" for each pair.
# Input: Two lists list1, list2 (same length)
# Output: List of strings
# Example:
# Input: [1, 2, 3], [1, 5, 3] → Output: ["Equal", "Different", "Equal"]

def compare_lst(lst1, lst2):
    lst = []
    for x, y in zip(lst1, lst2):
        if x == y:
            lst.append("Equal")
        else:
            lst.append("Different")
    return lst

compare_lst([1, 2, 3], [1, 5, 3])
```

```python
# Extract indices of even numbers
# Statement: Return all indices where the element is even.
# Input: List of integers arr
# Output: List of indices
# Example:
# Input: [1, 4, 3, 8, 5] → Output: [1, 3]
# Input: [1, 3, 5] → Output: []

def even_idx(lst):
    even_idx_lst = []
    for i, x in enumerate(lst):
        if x % 2 == 0:
            even_idx_lst.append(i)
    return even_idx_lst

even_idx([1, 4, 3, 8, 5])
```

```python
# Cumulative sum
# Statement: Create a new list where each element is the cumulative sum up to that index.
# Input: List of numbers arr
# Output: List of cumulative sums
# Example:
# Input: [1, 2, 3, 4] → Output: [1, 3, 6, 10]
# Input: [10, -5, 3] → Output: [10, 5, 8]

def cum_sum(lst):
    new_lst = []
    val = 0
    for i in lst:
        val += i
        new_lst.append(val)
    return new_lst

cum_sum([10, -5, 3])



```

```python
# Count character frequencies
# Statement: Count how many times each character appears in a string. Return a dictionary.
# Input: String text
# Output: Dictionary mapping character to count
# Example:
# Input: "hello" → Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}
# Input: "aabb" → Output: {'a': 2, 'b': 2}

def char_freq(word):
    dic = {}
    for i in word:
        if i not in list(dic.keys()):
            dic.update({i : 1})
        else:
            dic.update({i : dic[i] + 1})
    return dic

char_freq("aabbcdd")

            
```

```python
# Find pairs with sum
# Statement: Find all pairs of indices (i, j) 
# where i < j and arr[i] + arr[j] equals target.
# Input: List arr, integer target
# Output: List of tuples (i, j)
# Example:
# Input: [2, 7, 11, 15], 9 → Output: [(0, 1)] (2+7=9)
# Input: [1, 2, 3, 4], 5 → Output: [(0, 3), (1, 2)]

def sum_pairs(lst, num):
    out = []
    for i, x in enumerate(lst):
        for j, y in enumerate(lst):
            if x + y == num:
                out.append((i, j))
                
    return out

sum_pairs([1, 2, 3, 4], 5)
        
```

```python
# Rotate list elements
# Statement: Rotate list elements k positions to the right. Use a loop (not slicing).
# Input: List arr, integer k
# Output: Rotated list
# Example:
# Input: [1, 2, 3, 4, 5], 2 → Output: [4, 5, 1, 2, 3]
# Input: ['a', 'b', 'c'], 1 → Output: ['c', 'a', 'b']

def rotate(lst, num):
    new_lst = []
    for i, x in enumerate(lst):
        print(i, x)
        if i > num:
            new_lst.append(x)
    for i in lst:
        if i not in new_lst:
            new_lst.append(i)
    return new_lst

rotate(['a', 'b', 'c'], 1)

```
