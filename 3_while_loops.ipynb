```python
# Countdown
# Statement: Print numbers from n down to 1, then "Done!".
# Input: Single integer n
# Output: Print each number on a new line, then "Done!"
# Example:
# Input: 3 → Output: 3, 2, 1, Done!

def countdown(n):
    while n > 0:
        print(n)
        n -= 1
    print("Done")

countdown(5)
```

```python
# Find first power of 2 greater than n
# Statement: Find the smallest power of 2 that is greater than n.
# Input: Single integer n
# Output: Integer (power of 2)
# Example:
# Input: 10 → Output: 16 (2⁴)
# Input: 100 → Output: 128 (2⁷)

def pow_of_2(n):
    num = 1
    out = 0
    while out < n:
        cal = 2**num
        out = cal
        num += 1
    return out


pow_of_2(100)
```

```python
# Print digits of a number
# Statement: Print each digit of a number (from right to left, one per line).
# Input: Single integer n
# Output: Print each digit on a new line
# Example:
# Input: 1234 → Output: 4, 3, 2, 1

def print_rev(num):
    x = len(str(num))
    while x != 0:
        print(x)
        x -= 1

print_rev(1234)
```

# Medium problems

```python
# Collatz sequence length
# Statement: Calculate how many steps it takes for a number to reach 1 using Collatz: if even, divide by 2; if odd, multiply by 3 and add 1.
# Input: Single integer n
# Output: Integer (number of steps)
# Example:
# Input: 6 → Output: 8 (6→3→10→5→16→8→4→2→1)

def collatz(n):
    x = n
    count = 0
    while x != 1:
        if x % 2 == 0:
            x = x / 2
        else:
            x = (x * 3) + 1
        count += 1
        print(int(x))
    return count

collatz(6)
```

```python
# Find GCD using Euclidean algorithm
# Statement: Find GCD of two numbers using while loop 
# (Euclidean algorithm: while b != 0, set a, b = b, a % b).
# Input: Two integers a, b
# Output: Integer GCD
# Example:
# Input: 48, 18 → Output: 6
# Input: 17, 13 → Output: 1

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(gcd(48, 18))
```

```python
# Remove duplicates from list (in-place with while)
# Statement: Remove duplicate elements from a list using a while loop. Modify the original list.
# Input: List arr
# Output: Modified list (duplicates removed)
# Example:
# Input: [1, 2, 2, 3, 3, 3, 4] → Output: [1, 2, 3, 4]

def rem_dup(lst):
    new_lst = []
    x = 0
    while x < len(lst):
        if lst[x] not in new_lst:
            new_lst.append(lst[x])
        x += 1
    return new_lst

rem_dup([1, 2, 2, 3, 3, 3, 4])
```

# Hard problems

```python
# Binary representation
# Statement: Convert a number to binary representation (as a string) using a while loop.
# Input: Single integer n
# Output: String binary representation
# Example:
# Input: 10 → Output: "1010"
# Input: 0 → Output: "0"
def bin_rep(num):
    return bin(num)[2:]

bin_rep(0)
```

```python
# Factorial calculation
# Statement: Calculate factorial of n using a while loop.
# Input: Single integer n
# Output: Integer factorial
# Example:
# Input: 5 → Output: 120 (5! = 5×4×3×2×1)

def cal_factorial(num):
    x = num
    fact = 1
    while x != 0:
        fact *= x
        x -= 1
    return fact

cal_factorial(5)
```

```python
# Fibonacci sequence up to n
# Statement: Generate Fibonacci numbers up to (but not exceeding) n.
# Input: Single integer n
# Output: List of Fibonacci numbers
# Example:
# Input: 20 → Output: [0, 1, 1, 2, 3, 5, 8, 13]

def fib(n):
    
    a = 0
    b = 1
    lst = [a, b]
    while lst[-1] <= n:
        a, b = b, a + b
        if b >= n:
            break
        lst.append(b)
    return lst

fib(20)
    

```

```python
# Reverse digits of a number
# Statement: Reverse the digits of a number (e.g., 1234 → 4321).
# Input: Single integer n
# Output: Integer (reversed)
# Example:
# Input: 1234 → Output: 4321
# Input: 100 → Output: 1 (leading zeros don't count)

def rev_dig(num):
    num = str(num)
    x = len(str(num))
    rev_num = ""
    while x > 0:
        rev_num += num[x-1]
        x -= 1
    return int(rev_num)

rev_dig(1234)

```

```python
# Check if number is palindrome
# Statement: Check if a number reads the same forwards and backwards.
# Input: Single integer n
# Output: Boolean
# Example:
# Input: 121 → Output: True
# Input: 123 → Output: False

def palindrom(n):
    n = str(n)
    x = len(str(n))
    rev = ""
    while x > 0:
        rev += n[x - 1]
        x -= 1
    if rev == n:
        return True
    return False

palindrom(123)
```
